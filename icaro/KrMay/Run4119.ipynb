# Lifetime calculation

```python
fid = full
fid = fid[fid.R < 100] # Fiducialize in order to reduce geometrical effects
print("Ratio:", len(fid)/len(full))
fid.head()
```

# S2 energy fiducialized

```python
hist(fid.S2e, 50, range=Erange_raw)
labels("S2 energy (pes)", "Entries", "Fidualized energy spectrum")
save("S2energy_fiducial")

hist2d(fid.Z, fid.S2e, (50, 50), range=(Zrange, Erange_raw))
sel     = coref.in_range(fid.S2e.values, 11300 - 15*fid.Z.values, np.inf) # remove low E background
x, y, _ = fitf.profileX(fid.Z[sel], fid.S2e[sel], 30, xrange=Zrange, yrange=Erange_raw)
plt.plot(x, y, profOpt)
plt.plot(fid.Z.values, 11300 - 15*fid.Z.values, "k-")


seed = np.max(y), (x[15] - x[5])/np.log(y[15]/y[5])
f    = fitf.fit(fitf.expo, x, y, seed, fit_range=Zrange_LT)

plt.plot(x, f.fn(x), fitOpt)
print("Energy at z=0 = {:.1f} +- {:.1f}".format( f.values[0], f.errors[0]))
print("Lifetime      = {:.1f} +- {:.1f}".format(-f.values[1], f.errors[1]))
print("Chi2          = {:.2f}          ".format(f.chi2))

labels("Drift time ($\mu$s)", "S2 energy (pes)", "Fiducialized energy spectrum vs Z")
save("S2energyvsZ_fiducial")

Zcorr = corrf.LifetimeCorrection(-f.values[1], f.errors[1])
```

# Correction by attachment

```python
E = fid.S2e.values * Zcorr(fid.Z.values).value

y, x, _ = hist(E, 100, range=Erange_zcorr)

fun  = lambda x, *args: fitf.gauss  (x, *args[ :3])
seed = 1e3, 12000, 100
low  =  10, 11000,  50
upp  = 1e8, 13000, 900
f    = fitf.fit(fun, x, y, seed, fit_range=(11000, 13000), bounds=(low, upp))
print(f.values)


plt.plot(x, f.fn(x), fitOpt)
plt.plot(x, fitf.gauss(x, *f.values),  fitOpt)
plt.text(Erange_zcorr[0], 1*max(y)/4, gausstext(f.values, 41.5), color=fitOpt)

labels("S2 energy (pes)", "Entries", "Z-corrected energy")
save("S2energyCorr_fiducial")

hist2d(fid.Z, E, (50, 100), range=(Zrange, Erange_zcorr))
x, y, _ = fitf.profileX(fid.Z, E, 30, xrange=Zrange, yrange=Erange_zcorr)
plt.plot(x, y, profOpt)

f = fitf.fit(fitf.expo, x, y, (1e4, -800))
plt.plot(x, f.fn(x), fitOpt)

print(f.values)
labels("Drift time ($\mu$s)", "S2 energy (pes)", "Z-corrected energy vs Z")
save("S2energyCorrvsZ_fiducial")
```

# Correction by attachment and position

```python
Ecorr  = Kr.S2e.values * Zcorr(Kr.Z.values).value * XYcorr(Kr.X.values, Kr.Y.values).value

y, x, _ = hist(Ecorr, 80, Erange_xycorr)

fun  = fitf.gauss
seed = 1e3, 12000, 100
low  =  10, 11000,  50
upp  = 1e8, 13000, 900
f    = fitf.fit(fun, x, y, seed, fit_range=(11500, 13000), bounds=(low, upp), sigma=y**0.5)
print("Fit values:", f.values)
print("Fit errors:", f.errors)
print("Chi2      :", f.chi2)


plt.plot(x, f.fn(x), fitOpt)
plt.plot(x, fitf.gauss(x, *f.values),  fitOpt)
plt.text(Erange_xycorr[0], 1*max(y)/4, gausstext(f.values, 41.5), color=fitOpt)
labels("E (pes)", "Entries", "XYZ-corrected energy")
save("EcorrZXY")

_, _, cb = hist2d_profile(Kr.X.values, Kr.Y.values, Ecorr, 50, 50, XYrange, XYrange)
cb.set_label("E (pes)")
labels("x (mm)", "y (mm)", "XYZ-corrected energy vs XY")
save("EvsXY")

_, _, cb = hist2d_profile(Kr.R.values, Kr.Phi.values, Ecorr, 50, 50, Rrange, Prange)
cb.set_label("E (pes)")
labels("r (mm)", "$\phi$ (rad)", "XYZ-corrected energy vs R$\phi$")
save("EvsRPhi")

hist2d(Kr.R.values, Ecorr, (50, 50), range=(Rrange, Erange_xycorr))
x, y, _ = fitf.profileX(Kr.R.values, Ecorr, 50, Rrange, Erange_xycorr)
plt.plot(x, y, profOpt)
labels("r (mm)", "S2 energy (pes)", "XYZ-corrected energy vs R")
save("EvsR")

hist2d(Kr.Phi.values, Ecorr, (50, 50), range=(Prange, Erange_xycorr))
x, y, _ = fitf.profileX(Kr.Phi.values, Ecorr, 50, xrange=Prange, yrange=Erange_xycorr)
plt.plot(x, y, profOpt)
labels("phi (rad)", "S2 energy (pes)", "XYZ-corrected energy vs $\phi$")
save("EvsPhi")
```
